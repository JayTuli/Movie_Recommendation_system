```python
# can create a helper function
import ast
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
def convert3(obj):
  L =[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter=+1
    else:
        break
  return L
```

```python
def fetch_director(obj):
  L =[]
  counter = 0
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =  PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
---
---
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
 from sklearn.feature_extraction.text import CountVectorizer
 cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

to find step by step distance of movies with other 4806 movies
total distance is 4806 X 4806

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]     # SIMILARITY SCORE OF FIRST MOVIE AND EVERY MOVIES (distance)
```

we will call enumerate function

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['original_title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].original_title)
```

```python
recommend('Batman Begins')
```
